```python
import numpy as np

heart_rate_data = np.random.randint(60, 110, 1440)
print(heart_rate_data)
```

```python
from scipy.signal import butter, filtfilt

def lowpass_filter(data, cutoff, fs, order=5):
    nyquist = 0.5 * fs 
    normal_cutoff = cutoff / nyquist
    b, a = butter(order, normal_cutoff)
    y = filtfilt(b, a, data)
    return y
cutoff = 0.1
fs = 1 
order = 3  

smoothed_data = lowpass_filter(heart_rate_data, cutoff, fs, order)
print(smoothed_data)
```

```python
def compute_hourly_averages(data):
    hourly_averages = []
    for i in range(0, len(data), 60):
        hourly_avg = np.mean(data[i:i+60])
        hourly_averages.append(hourly_avg)
    return hourly_averages

hourly_averages = compute_hourly_averages(smoothed_data)
print(hourly_averages)
```

```python
import matplotlib.pyplot as plt

time_minutes = np.arange(0, len(heart_rate_data))
time_hours = np.arange(0, len(hourly_averages))
for i in range(0, len(hourly_averages)):
    time_hours[i] = time_hours[i]*60

print(time_minutes)
print(time_hours)
plt.figure(figsize=(10, 6))

plt.plot(time_minutes, heart_rate_data, color="r", label="Original Heart")
plt.plot(time_minutes, smoothed_data, color="b", label="Smoothed Data")
plt.plot(time_hours, hourly_averages, color="g", label="Hourly Average")
plt.show()
```
